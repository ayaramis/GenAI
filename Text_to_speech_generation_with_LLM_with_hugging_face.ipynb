```python
from transformers import pipeline
```

```python
text = "Biotechnology is technology that utilizes biological systems, living organisms or parts of this to develop or create different products."

pipe = pipeline("text-to-speech", model="suno/bark-small", device="cuda")

output = pipe(text)
```

```python
output
```

```python
from IPython.display import Audio

Audio(output["audio"], rate=output["sampling_rate"])
```
